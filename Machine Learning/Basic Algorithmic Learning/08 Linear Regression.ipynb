```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
from collections import Counter
```

```python
%matplotlib inline
rcParams['figure.figsize']=5,4
sb.set_style('whitegrid')
```

```python
address="C:\Data_jupyter\enrollment_forecast.csv"
enroll=pd.read_csv(address)
enroll.columns=['year','roll','unem','hgrad','inc']
enroll.head()
```

```python
sb.pairplot(enroll)
```

```python
print(enroll.corr()) #Here (unem,hgrad) has low correlation than we can do linear reg
```

```python
enroll_data=enroll.iloc[:,[2,3]].values #unem,hgrad variable are seperated 
enroll_target=enroll.iloc[:,[1]].values 
enroll_data_names=['unem','hgrad']

X,y=scale(enroll_data,),enroll_target #scale the data
```

```python
#check the missing value
missing_values= X==np.NAN #check the missing values
X[missing_values==True] #Return array of missing values
```

```python
#No missing value so we can do linear reg

LinReg = LinearRegression(normalize=True)
LinReg.fit(X,y)
print(LinReg.score(X,y)) #This means how well the model performs in predicting the values.  always bw[0,1]
```
