```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats import spearmanr #Spearman Rank Corelation method
```

```python
address="C:\Data_jupyter\mtcars.csv"
cars=pd.read_csv(address)
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```python
sb.pairplot(cars)
```

```python
X=cars[['cyl','vs','am','gear']]
sb.pairplot(X)
```

### Spearman rank correaltion Coefficient
https://www.statisticshowto.datasciencecentral.com/spearman-rank-correlation-definition-calculate/

```python
cyl=cars['cyl']
am=cars['am']
vs=cars['vs']
gear=cars['gear']
#spearman rank corelation
spearmanr_coefficient, p_value=spearmanr(cyl,vs)
print("Spearman rank correaltion Coefficient: {}".format(spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value=spearmanr(cyl,am)
print("Spearman rank correaltion Coefficient: {}".format(spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value=spearmanr(cyl,gear)
print("Spearman rank correaltion Coefficient: {}".format(spearmanr_coefficient))
```

```python
#Chi Square TEST
table=pd.crosstab(cyl,am)
from scipy.stats import chi2_contingency
chi2, p, dof, expected =chi2_contingency(table.values)
print("Chi Square statistics:",chi2)
print("P-value",p)
```

```python
table=pd.crosstab(cyl,vs)
chi2, p, dof, expected =chi2_contingency(table.values)
print("Chi Square statistics:",chi2)
print("P-value",p)
```
